# L2 Cache Batch Test Results - Line Chart Analysis

**Generated:** 2025-11-17 17:43:12

**Data Source:** l2_batch_summary_20251117_173007.csv

**Total Tests:** 7


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 8)
plt.rcParams['font.size'] = 10
%matplotlib inline
```

```python
# Load batch test summary
df = pd.read_csv('l2_batch_summary_20251117_173007.csv')

print(f'Total test configurations: {len(df)}')
print(f'Columns: {list(df.columns)}')
display(df.head(10))

# Convert numeric columns
df['l2_on_tp_tps'] = pd.to_numeric(df['l2_on_tp_tps'], errors='coerce')
df['l2_off_tp_tps'] = pd.to_numeric(df['l2_off_tp_tps'], errors='coerce')
df['l2_on_ap_qps'] = pd.to_numeric(df['l2_on_ap_qps'], errors='coerce')
df['l2_off_ap_qps'] = pd.to_numeric(df['l2_off_ap_qps'], errors='coerce')
df['tp_change_pct'] = pd.to_numeric(df['tp_change_pct'], errors='coerce')
df['ap_change_pct'] = pd.to_numeric(df['ap_change_pct'], errors='coerce')
```

## Chart 1: TP Performance (TPS) by TP:AP Ratio

Shows how TP performance changes with different TP:AP ratios:


```python
# Prepare data for TP performance chart
df_chart = df.copy()
df_chart = df_chart.sort_values('tp_ap_ratio')

fig, ax = plt.subplots(figsize=(14, 6))

# Plot L2 ON and L2 OFF lines
x_pos = range(len(df_chart))
ax.plot(x_pos, df_chart['l2_off_tp_tps'], marker='o', linewidth=2, markersize=8, 
        label='L2 OFF', color='#e74c3c', linestyle='--')
ax.plot(x_pos, df_chart['l2_on_tp_tps'], marker='s', linewidth=2, markersize=8, 
        label='L2 ON', color='#2ecc71')

# Customize x-axis
ax.set_xticks(x_pos)
ax.set_xticklabels(df_chart['tp_ap_ratio'], rotation=45, ha='right')
ax.set_xlabel('TP:AP Ratio', fontsize=12, fontweight='bold')
ax.set_ylabel('TPS (Transactions Per Second)', fontsize=12, fontweight='bold')
ax.set_title('TP Performance: L2 ON vs L2 OFF by TP:AP Ratio', fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='best', fontsize=11)
ax.grid(True, alpha=0.3, linestyle=':')

# Add value labels on points
for i, (off_val, on_val) in enumerate(zip(df_chart['l2_off_tp_tps'], df_chart['l2_on_tp_tps'])):
    ax.annotate(f'{int(off_val)}', (i, off_val), textcoords='offset points', 
                xytext=(0,10), ha='center', fontsize=8, color='#e74c3c')
    ax.annotate(f'{int(on_val)}', (i, on_val), textcoords='offset points', 
                xytext=(0,-15), ha='center', fontsize=8, color='#2ecc71')

plt.tight_layout()
plt.show()
```

## Chart 2: AP Performance (QPS) by TP:AP Ratio

Shows how AP performance changes with different TP:AP ratios:


```python
# Prepare data for AP performance chart
fig, ax = plt.subplots(figsize=(14, 6))

# Plot L2 ON and L2 OFF lines
ax.plot(x_pos, df_chart['l2_off_ap_qps'], marker='o', linewidth=2, markersize=8, 
        label='L2 OFF', color='#e74c3c', linestyle='--')
ax.plot(x_pos, df_chart['l2_on_ap_qps'], marker='s', linewidth=2, markersize=8, 
        label='L2 ON', color='#2ecc71')

# Customize x-axis
ax.set_xticks(x_pos)
ax.set_xticklabels(df_chart['tp_ap_ratio'], rotation=45, ha='right')
ax.set_xlabel('TP:AP Ratio', fontsize=12, fontweight='bold')
ax.set_ylabel('QPS (Queries Per Second)', fontsize=12, fontweight='bold')
ax.set_title('AP Performance: L2 ON vs L2 OFF by TP:AP Ratio', fontsize=14, fontweight='bold', pad=20)
ax.legend(loc='best', fontsize=11)
ax.grid(True, alpha=0.3, linestyle=':')

# Add value labels on points
for i, (off_val, on_val) in enumerate(zip(df_chart['l2_off_ap_qps'], df_chart['l2_on_ap_qps'])):
    ax.annotate(f'{int(off_val)}', (i, off_val), textcoords='offset points', 
                xytext=(0,10), ha='center', fontsize=8, color='#e74c3c')
    ax.annotate(f'{int(on_val)}', (i, on_val), textcoords='offset points', 
                xytext=(0,-15), ha='center', fontsize=8, color='#2ecc71')

plt.tight_layout()
plt.show()
```

## Chart 3: Performance Improvement Percentage

Shows the percentage change for TP and AP workloads:


```python
# Performance improvement chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# TP improvement
ax1.plot(x_pos, df_chart['tp_change_pct'], marker='o', linewidth=2, markersize=8, 
         color='#3498db', label='TP Change %')
ax1.axhline(y=0, color='black', linestyle='--', linewidth=1, alpha=0.5)
ax1.set_xticks(x_pos)
ax1.set_xticklabels(df_chart['tp_ap_ratio'], rotation=45, ha='right')
ax1.set_xlabel('TP:AP Ratio', fontsize=11, fontweight='bold')
ax1.set_ylabel('Change Percentage (%)', fontsize=11, fontweight='bold')
ax1.set_title('TP Performance Change (L2 ON vs L2 OFF)', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3, linestyle=':')
ax1.legend()

# Add value labels
for i, val in enumerate(df_chart['tp_change_pct']):
    ax1.annotate(f'{val:.1f}%', (i, val), textcoords='offset points', 
                xytext=(0,10), ha='center', fontsize=8)

# AP improvement
ax2.plot(x_pos, df_chart['ap_change_pct'], marker='s', linewidth=2, markersize=8, 
         color='#9b59b6', label='AP Change %')
ax2.axhline(y=0, color='black', linestyle='--', linewidth=1, alpha=0.5)
ax2.set_xticks(x_pos)
ax2.set_xticklabels(df_chart['tp_ap_ratio'], rotation=45, ha='right')
ax2.set_xlabel('TP:AP Ratio', fontsize=11, fontweight='bold')
ax2.set_ylabel('Change Percentage (%)', fontsize=11, fontweight='bold')
ax2.set_title('AP Performance Change (L2 ON vs L2 OFF)', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3, linestyle=':')
ax2.legend()

# Add value labels
for i, val in enumerate(df_chart['ap_change_pct']):
    ax2.annotate(f'{val:.1f}%', (i, val), textcoords='offset points', 
                xytext=(0,10), ha='center', fontsize=8)

plt.tight_layout()
plt.show()
```

## Chart 5: Combined Throughput Comparison

以负载情况（TP:AP比例）为横坐标，显示TP和AP在L2 ON/OFF下的吞吐量对比：


```python
import matplotlib.pyplot as plt
import pandas as pd

# Combined throughput chart: TP and AP for L2 ON/OFF
fig, ax = plt.subplots(figsize=(16, 8))

# Plot all four lines
ax.plot(x_pos, df_chart['l2_off_tp_tps'], marker='o', linewidth=2.5, markersize=10, 
         label='TP L2 OFF', color='#e74c3c', linestyle='--', alpha=0.8)
ax.plot(x_pos, df_chart['l2_on_tp_tps'], marker='s', linewidth=2.5, markersize=10, 
         label='TP L2 ON', color='#2ecc71', linestyle='-', alpha=0.8)
ax.plot(x_pos, df_chart['l2_off_ap_qps'], marker='^', linewidth=2.5, markersize=10, 
         label='AP L2 OFF', color='#f39c12', linestyle='--', alpha=0.8)
ax.plot(x_pos, df_chart['l2_on_ap_qps'], marker='D', linewidth=2.5, markersize=10, 
         label='AP L2 ON', color='#3498db', linestyle='-', alpha=0.8)

# Customize x-axis
ax.set_xticks(x_pos)
ax.set_xticklabels(df_chart['tp_ap_ratio'], rotation=45, ha='right', fontsize=11)
ax.set_xlabel('负载情况 (TP:AP Ratio)', fontsize=13, fontweight='bold')
ax.set_ylabel('吞吐量 (Throughput)', fontsize=13, fontweight='bold')
ax.set_title('L2 Cache性能对比: TP和AP负载在不同配置下的吞吐量', fontsize=15, fontweight='bold', pad=20)
ax.legend(loc='best', fontsize=11, framealpha=0.9)
ax.grid(True, alpha=0.3, linestyle=':')

# Add value labels on key points (only for non-zero values)
for i in range(len(df_chart)):
    # TP L2 OFF
    if not pd.isna(df_chart.iloc[i]['l2_off_tp_tps']) and df_chart.iloc[i]['l2_off_tp_tps'] > 0:
        ax.annotate(f"{int(df_chart.iloc[i]['l2_off_tp_tps'])}", 
                    (i, df_chart.iloc[i]['l2_off_tp_tps']), 
                    textcoords='offset points', xytext=(0,12), ha='center', 
                    fontsize=8, color='#e74c3c', fontweight='bold')
    # TP L2 ON
    if not pd.isna(df_chart.iloc[i]['l2_on_tp_tps']) and df_chart.iloc[i]['l2_on_tp_tps'] > 0:
        ax.annotate(f"{int(df_chart.iloc[i]['l2_on_tp_tps'])}", 
                    (i, df_chart.iloc[i]['l2_on_tp_tps']), 
                    textcoords='offset points', xytext=(0,-15), ha='center', 
                    fontsize=8, color='#2ecc71', fontweight='bold')
    # AP L2 OFF
    if not pd.isna(df_chart.iloc[i]['l2_off_ap_qps']) and df_chart.iloc[i]['l2_off_ap_qps'] > 0:
        ax.annotate(f"{int(df_chart.iloc[i]['l2_off_ap_qps'])}", 
                    (i, df_chart.iloc[i]['l2_off_ap_qps']), 
                    textcoords='offset points', xytext=(0,12), ha='center', 
                    fontsize=8, color='#f39c12', fontweight='bold')
    # AP L2 ON
    if not pd.isna(df_chart.iloc[i]['l2_on_ap_qps']) and df_chart.iloc[i]['l2_on_ap_qps'] > 0:
        ax.annotate(f"{int(df_chart.iloc[i]['l2_on_ap_qps'])}", 
                    (i, df_chart.iloc[i]['l2_on_ap_qps']), 
                    textcoords='offset points', xytext=(0,-15), ha='center', 
                    fontsize=8, color='#3498db', fontweight='bold')

plt.tight_layout()
plt.show()
```
